```python
def probPlt(data,i, L = None):
    
    if L == None:
        L = 1
    
    i = int(i)

    length = len(data[0])
    
    dz = L/length
    
    plt.plot( hb.initL(L,dz)['prob'], '-k', label='Vacuum' )
    plt.plot( 2* data[i], '-b', label = 'Forward')    
    plt.plot( 2* data[i+1][::-1], '--r', label='Back')
    plt.ylim([-1.1,1.1])
    plt.legend()
    plt.show()
```

```python
prob1000e5Data = np.genfromtxt('prob-1000-e-5.csv', delimiter=',')
prob1000e4R5Data = np.genfromtxt('prob-1000-e-4-range-5.csv', delimiter=',')

prob1000e5RevisedData = np.genfromtxt('revised/prob-1000-e-5-test.csv', delimiter=',')
prob1000e5RevisedData = prob1000e5RevisedData[:14]

probtest = np.genfromtxt('prob-test.csv', delimiter=',')
```
